```python
# 두 데이터프레임 병합 (articleID를 기준으로)
merged_df = pd.merge(view_log, article_info, on='articleID', how='inner')

# 병합된 데이터 확인
print(merged_df.info())
merged_df.head()
```

```python
# 결측치 확인
print(merged_df.isnull().sum())
```

```python
# 결측치 처리 
merged_df['userCountry_y'].fillna('Unknown', inplace=True)
merged_df['userRegion_y'].fillna('Unknown', inplace=True)

# 불필요한 열 제거 (article_info.csv에서 중복된 userID, userCountry, userRegion 제거)
merged_df.drop(columns=['userCountry_y', 'userRegion_y'], inplace=True)
```

```python
merged_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 기사 형식별 분포 시각화
plt.figure(figsize=(10, 6))
sns.countplot(data=merged_df, x='Format', order=merged_df['Format'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of Articles by Format')
plt.show()

# 언어별 기사 수 시각화
plt.figure(figsize=(10, 6))
sns.countplot(data=merged_df, x='Language', order=merged_df['Language'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of Articles by Language')
plt.show()

# 지역별 사용자 수 시각화
plt.figure(figsize=(10, 6))
sns.countplot(data=merged_df, x='userRegion_x', order=merged_df['userRegion_x'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of Users by Region')
plt.show()

# 국가별 사용자 수 시각화
plt.figure(figsize=(10, 6))
sns.countplot(data=merged_df, x='userCountry_x', order=merged_df['userCountry_x'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of Users by Country')
plt.show()
```
